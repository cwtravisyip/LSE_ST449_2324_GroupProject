```python
import os
import time
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/19DEC2023_AMS.csv")
```

```python
df
```

```python
G = nx.DiGraph()

# Adding nodes and edges based on the flight data
for index, row in df.iterrows():
    G.add_edge(row['orig'], row['dest'], departure=row['time_sch'])
```

```python
plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title("Flight Network Graph")
plt.show()
```
